# How to easily download images using Chrome command line

Here's a little hack for downloading images on a page (not necessary for resizing)

First, go to this page http://vision.stanford.edu/aditya86/ImageNetDogs/

Then open the console. In the console, run this code to get all the image elements:
```
dogs = document.querySelectorAll('a img')
```


Then run this code to get all the src urls for the images:

```
urls = []

for (var i = 0; i < dogs.length; i ++ ){
  urls.push(dogs[i].src)
}
```


Then run `JSON.stringify(urls)` to turn it into a string, and copy it using the 'copy' button.

# How to resize images

````python
# first run
# pip3 install Pillow
# to install PIL, the "python imaging library"

from PIL import Image
````

````python
import glob

# get all our image names
image_names = glob.glob('./dogs/*.jpg')

# we also need to make a folder to save the resized files into

! mkdir resized
````

````python
image_names
````

````python
for image_name in image_names:
    image = Image.open(image_name)
    # the thumbnail command will maintain the aspect ratio 
    # but make sure the width/height is not larger than
    # the number specified - 60 pixels in this case
    image.thumbnail((60,60))
    
    # now we need the name of our image
    image_name = image_name.split('/')[-1]
    image.save('resized/' + image_name)
````
